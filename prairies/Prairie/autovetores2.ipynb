```python
import matplotlib.pyplot as plt
import numpy as np

# Gerando dados para o exemplo
np.random.seed(0)
x = np.random.randn(50)
y = 2 * x + np.random.randn(50)

# Plotando o gráfico de dispersão
plt.figure(figsize=(8, 6))
plt.scatter(x, y, label='Dados', color='blue')

# Plotando o autovetor
autovetor_x = 0.61
autovetor_y = -0.78
plt.arrow(0, 0, autovetor_x, autovetor_y, head_width=0.1, head_length=0.1, fc='red', ec='red')
plt.text(0.5 * autovetor_x, 0.5 * autovetor_y, 'Autovetor', color='red')

plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Gráfico de Dispersão com Autovetor')
plt.legend()
plt.grid(True)
plt.show()
```
